Adding the new column success which consider a company that is closed a fail and a success if was acquired and is active or if is top500 and active

```python
def fix_closed(x):
    if x.startswith('a'):
        return 'active'
    else:
        return 'notactive'

df_new['closed_at'] = df_new['closed_at'].apply(fix_closed)
```

```python
def success_or_not(df_new):
    df1=df_new.copy()
    
    if (((df1['is_top500']==1)|(df1['is_acquired']==1)) & (df1['closed_at']=="active")):
        return 1
    elif (((df1['is_top500']==0) & (df1['is_acquired']==0)) | (df1['closed_at']=="notactive")):
        return 0
```

```python
df_new['success'] = df_new.apply(success_or_not,axis=1)
```
